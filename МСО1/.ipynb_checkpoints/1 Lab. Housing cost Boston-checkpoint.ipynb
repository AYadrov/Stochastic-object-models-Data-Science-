## Content

Кол-во экземпляров: 506

Информация об атрибутах:

 - CRIM: уровень преступности на душу населения по рассматриваемым округам
 - ZN: доля земель под жилую застройку, площадью более 25 000 кв. футов
 - INDUS: доля акров, не связанных с розничной торговлей, на город
 - CHAS: фиктивная переменная реки Чарльз (= 1, если участок ограничивает реку; 0 в противном случае)
 - NOX: концентрация оксидов азота (частей на 10 миллионов)
 - RM: среднее количество комнат в доме
 - AGE: доля занимаемых владельцами единиц, построенных до 1940 г.
 - DIS: взвешенные расстояния до пяти бостонских центров занятости
 - RAD: индекс доступности радиальных магистралей
 - TAX: полная ставка налога на имущество за 10 000 долларов.
 - PTRATIO: соотношение учеников и учителей по городам
 - B: 1000 (Bk - 0,63) ^ 2, где Bk - доля темнокожих по городам
 - LSTAT: % более низкий статус населения
 - MEDV: Средняя стоимость домов, занимаемых владельцами в 1000 долларов США.

```python
sns.jointplot(x=data['CRIM'], y=data['LSTAT'], kind="reg")
```

#### Уровень преступности связан с более низким статусом населения

```python
plt.figure(figsize=(15,8))
ax = sns.boxplot(x=data['CRIM'].round(), y=data["MEDV"], palette='Set2')
ax.set_xlabel('Crime rate (rounded)')
ax.set_ylabel('Cost if dwelling')
ax.set_title("The cost of apartments depending on the crime rate")
```

#### Как можно видеть, самое дорогое жилье находится в районах с самой низкой преступностью. Также стоимость постепенно идёт на спад при увеличении преступности

```python
ax = sns.jointplot(x=data['NOX'], y=data['MEDV'], kind="scatter")
```

#### Большая концентрация оксидов азота имеет отрицательное воздействие на организм человека. Можно сделать вывод, что наиболее дорогие жилища, выше среднего показателя(22 тыс. $) находятся в основном не в центрах активности азота. Следовательно, этот фактор имеет небольшое влияение на конечную стоимость. 

```python
plt.figure(figsize=(10,7))
data['B'].hist()
plt.xlabel('Доля темнокожих')
plt.ylabel('Кол-во городов')
plt.show()
```

#### Доля темнокожих в основном одинаковая в городах близ Бостона

```python
ax = sns.jointplot(x=data['CRIM'], y=data['B'], kind="kde")
black_level = data['B'].mean()
low_black = data[data['B']<black_level]
high_black = data[data['B']>=black_level]
low_black = low_black['CRIM'].mean()
high_black = high_black['CRIM'].mean()
print(black_level)
print('Средний уровень преступности, где доля темнокожего населения меньше среднего = ' + str(low_black))
print('Средний уровень преступности, где доля темнокожего населения выше среднего = ' + str(high_black))
```

#### Можно видеть, что доля темнокожего населения не влияет на криминальный уровень.

```python
ax = sns.jointplot(x=data['LSTAT'], y=data['B'], kind="kde")
black_level = data['B'].mean()
low_black = data[data['B']<black_level]
high_black = data[data['B']>=black_level]
low_black = low_black['LSTAT'].mean()
high_black = high_black['LSTAT'].mean()
print('Средний процент населения с низким статусом, где доля темнокожего населения меньше среднего = ' + str(low_black))
print('Средний процент населения с низким статусом, где доля темнокожего населения выше среднего = ' + str(high_black))
```

#### Также стоит отметить, что доля темнокожих влияет обратнопропорционально на процент населения с низким статусом

#### Следовательно, где доля темнокожих будет больше, там будет более благополучный район в плане криминальной статистики и статуса людей

```python
ax = sns.jointplot(x=data['MEDV'], y=data['B'], kind="kde")
black_level = data['B'].mean()
low_black = data[data['B']<black_level]
high_black = data[data['B']>=black_level]
low_black = low_black['MEDV'].mean()
high_black = high_black['MEDV'].mean()
print('Средняя цена жилья, где доля темнокожего населения меньше среднего = ' + str(low_black) + " тыс. $")
print('Средняя цена жилья, где доля темнокожего населения выше среднего = ' + str(high_black) + " тыс. $")
```

#### Сделав прошлым шагом вывод, о том, что более благополучным будет считаться район, где темнокожего населения больше, мы видим, что в среднем и цена жилья выше. Значит, что процент темнокожих влияет на цену. Где больше темнокожих, тем больше цена
